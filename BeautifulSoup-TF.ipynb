```
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```
url = 'https://www.transfermarkt.es/leonardo-campana/profil/spieler/606872'
```

```
response = requests.get(url, headers={'User-Agent': 'Custom5'})
print(response.status_code)
```

```
player_data = response.text
soup = BeautifulSoup(player_data, 'html.parser')
```

```
grouped_data = soup.find('table', {'class': 'auflistung'})
```

```
name_player = str(grouped_data.find_all('td')[0].text.rstrip('<td>'))
birthdate_player = str(grouped_data.find_all('td')[1].text.rstrip(' <td>'))
age_player  = str(grouped_data.find_all('td')[3].text.rstrip('<td>'))
nationality_player  = str(grouped_data.find_all('td')[5].text.rstrip('')).replace('\n','').replace('\xa0','').replace(' ','')
position_player =str(grouped_data.find_all('td')[6].text.rstrip('<td>')).replace('\n','').replace(' ','')
foot_player = str(grouped_data.find_all('td')[7].text.rstrip('<td>')).replace('\n','').replace(' ','')
height_player = str(grouped_data.find_all('td')[4].text.rstrip('<td>')).replace('\n','').replace(' ','').replace('\xa0m','')
club_player = str(grouped_data.find_all('td')[9].text.rstrip('<td>')).replace('\n','')
```

```
player_df = pd.DataFrame({'Name': [name_player],
                          'Nationality': [nationality_player],
                          'Birthdate': [birthdate_player],
                          'Age': [age_player],
                          'Height': [height_player],
                          'Foot': [foot_player],
                          'Position': [position_player],
                          'Club': [club_player]
                            })
```
